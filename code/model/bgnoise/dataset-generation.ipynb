```python
import os
import random

from pydub import AudioSegment
import pandas as pd
```

```python
fgSound = AudioSegment.from_file("../../data/snsd/clean/clnsp0.wav")
bgSound = AudioSegment.from_file("../../data/snsd/noise/AirConditioner_1.wav")

bgSound -= 10

combined = fgSound.overlay(bgSound)

combined.export("noisy.wav", format='wav')
```

```python
def overlay(clean_file_path, noise_file_path, noise_level, filedir='.'):
    fgSound = AudioSegment.from_file(clean_file_path)
    bgSound = AudioSegment.from_file(noise_file_path)

    bgSound -= 10 if noise_level == 'high' else 20 if noise_level == 'medium' else 30

    combined = fgSound.overlay(bgSound)
    new_file_name = f"{clean_file_path.split('/')[-1].split('.')[0]}-{noise_level}.wav"

    combined.export(f"{filedir}/{new_file_name}", format='wav')
    return new_file_name
```

```python
overlay('../../data/snsd/clean/clnsp0.wav', '../../data/snsd/noise/AirConditioner_1.wav', 'low')
overlay('../../data/snsd/clean/clnsp0.wav', '../../data/snsd/noise/Typing_10.wav', 'medium')
overlay('../../data/snsd/clean/clnsp0.wav', '../../data/snsd/noise/AirConditioner_1.wav', 'high')
```

## Generating the dataset

```python
noise_folder_path = '../../data/snsd/noise'
clean_folder_path = '../../data/snsd/clean'
noise_files = [f'{noise_folder_path}/{file}' for file in os.listdir(noise_folder_path)]
clean_files = [f'{clean_folder_path}/{file}' for file in os.listdir(clean_folder_path)]
```

```python
dataset = [
    {
        'label': sound_level,
        'path': overlay(clean_file, random.choice(noise_files), sound_level, f'../../data/snsd/generated/noise_{sound_level}')
    }
    for sound_level in ['low', 'medium', 'high']
    for clean_file in clean_files
]
```
